```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('tracks.csv')
df_tracks.head()
```

```python
# Check null values and sum them for each column

null_counts = pd.isnull(df_tracks).sum()

print(null_counts)
```

```python
#determine the total rows datatypes and memory usage
df_tracks.info()
```

```python
#finding 10 least popular songs

sorted_df = df_tracks.sort_values('popularity', ascending=True).head(10)  
print(sorted_df)  
```

```python
#descriptive statistics

df_tracks.describe().transpose()
```

```python
#most popular songs greater than 90

most_popular = df_tracks.query('popularity>90' , inplace = False ).sort_values('popularity', ascending = False)
most_popular[:10]

```

```python
if 'release_date' in df_tracks.columns:
    # Ensure no missing values in the column (optional)
    if df_tracks['release_date'].isnull().any():
        print("Warning: 'release_date' contains missing values. Consider handling them before setting the index.")

    # Set 'release_date' as the index
    df_tracks.set_index("release_date", inplace=True)
    
    # Convert the index to datetime format
    df_tracks.index = pd.to_datetime(df_tracks.index, errors='coerce')

    # Print the first few rows to verify
    print(df_tracks.head())
else:
    print("The column 'release_date' does not exist in the DataFrame.")
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
#covert songs duration from miliseconds to only seconds 

df_tracks["duration"]= df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis=1)

df_tracks.duration.head()
```

```python
#determine the correaltion between variables
#drop non-numeric columns that are not relevant
non_numeric_cols = df_tracks.select_dtypes(include=['object']).columns
print("Non-numeric columns:", non_numeric_cols)

#If these columns are not relevant, drop them
df_tracks_cleaned = df_tracks.drop(columns=non_numeric_cols)

#Convert categorical data to numerical if needed

df_tracks_encoded = pd.get_dummies(df_tracks_cleaned, drop_first=True)

corr_df = df_tracks_encoded.corr(method="pearson")
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(
    corr_df,
    annot=True,
    fmt='.1g',
    vmin=-1,
    vmax=1,
    center=0,
    cmap="inferno",
    linewidths=1,
    linecolor="black"
)
heatmap.set_title("Correlation HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
plt.show()
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
#create regression between loudness and energy as they have positive correlation

plt.figure(figsize=(10, 6))
sns.regplot(data=sample_df, y="loudness", x="energy", color="c")
plt.title("Loudness vs Energy Correlation")
plt.show()

```

```python
#same process for popularity and acousticness as did above

plt.figure(figsize=(10, 6))
sns.regplot(data=sample_df, y="popularity", x="acousticness", color="b")
plt.title("Popularity vs Acousticness")
plt.show()
#negative correlation
```

Import new dataset for genre

```python
#import new datasets to analyze Genre of songs

df_genre = pd.read_csv("SpotifyFeatures.csv")
df_genre.head()
```

```python
null_counts = pd.isnull(df_genre).sum()

print(null_counts)
```

```python
df_genre.info()
```

```python
#duration of songs or different genres

# color palette for the bar plot
palette = sns.color_palette("rocket")
plt.figure(figsize=(12, 8))
sns.barplot(y="genre", x="duration_ms", data=df_genre, palette=palette)

plt.title("Duration of the Songs in Different Genres")
plt.xlabel("Duration in Milliseconds")
plt.ylabel("Genres")
plt.show()
```

```python
#top 5 geners by popularity

sns.set_style("darkgrid")
plt.figure(figsize=(10, 5))
famous = df_genre.sort_values("popularity", ascending=False).head(10)
sns.barplot(
    y="genre",
    x="popularity",
    data=famous,
    palette="Spectral"  #(colorful palette)
)

plt.title("Top 5 Genres by Popularity")
plt.show()
```
